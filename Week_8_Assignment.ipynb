2. How many total unique words are in the corpus? (Please feel free to define unique words in any interesting, defensible way).

```python
fdist1.hapaxes()

# words/tokens that occur once only
```

```python
len(fdist1.hapaxes())

# the hapaxes method provides an output of a list which contains a list of words that only occur once in the text
# in this case, there are 9002 unique words in this corpus
```

3. Taking the most common words, how many unique words represent half of the total words in the corpus?

```python
# I have calculated 9002 words to be 'unique' to the text/corpus - they occur only once
# so in this case, all these words combined don't represent half of the total words in the corpus
# more like 35%

9002/len(text1)
```

4.Identify the 200 highest frequency words in this corpus.

```python
# Below are the 200 highest frequency words in this corpus

data = list(sorted(fdist1.items(), key=lambda item: item[1]))[-200:]

data
```

5. Create a graph that shows the relative frequency of these 200 words.

```python
df = pd.DataFrame.from_records(data, columns=['Word', 'Count'])

df
```

```python
df.plot(kind = 'hist',
        x = 'Word',
        y = 'Count'
       )

plt.title('Word Frequency')
```
